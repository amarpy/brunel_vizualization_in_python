# Basic Brunel Charts

```python
#!pip install brunel
import pandas as pd
import brunel
```

## Data Loading and Preprocessing

```python
superstore = pd.read_excel("datasets/superstore.xls")
```

```python
superstore.columns
```

```python
new_columns = []
for name in superstore.columns:
    name = name.replace(" ", "_").replace("-", "_")
    new_columns.append(name)
    
superstore.columns = new_columns
```

```python
superstore.columns
```

## Stacked Bar Chart

```python
%brunel data("superstore") bar x(Category) y(Sales) stack color(Region) sum(Sales) sort(Sales) tooltip(#all) :: width=800, height=300
```

## Scatter plot

```python
%brunel data("superstore") x(Sales) y(Profit) color(Category) size(Quantity) tooltip(#all) :: width = 800, height = 600
```

## Chord Chart

```python
%brunel data("superstore") x(Category) y(Ship_Mode) chord size(Quantity) sum(Quantity) color(Category) tooltip(#all) :: width = 600, height = 600 
```

## Tree Map

```python
%brunel data("superstore") treemap x(Region, Category, Sub_category) color(Profit:[orange, **, blue]) sum(Profit) size(Sales) sum(Sales) tooltip(#all):: width=800, height=600
```
